```
# Stock Market Trend Prediction (Next 5-Candles Trend)

# Step 1: Import Libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from google.colab import drive
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

drive.mount("/content/drive")
# Step 2: Load Dataset
# Using raw string for Windows path
df = pd.read_csv(r'/content/drive/MyDrive/Assignment/large_32 (2).csv')
# Alternatively, you can use forward slashes:
# df = pd.read_csv('C:/Users/VICTUS/Stock/large_32 (2).csv')
df['Date Time'] = pd.to_datetime(df['Date Time'])
df = df.sort_values('Date Time').reset_index(drop=True)

# Step 3: Inspect Dataset
print(df.head())
print(df.info())
print(df.isnull().sum())

# Optional: Fill or drop missing values
df.fillna(method='ffill', inplace=True)

# Step 4: Label Generation for Next 5-Candles Trend
def generate_labels(data):
    labels = []
    for i in range(len(data) - 5):
        next_candles = data[i+1:i+6]
        if all(next_candles['Close'] > next_candles['Open']):
            labels.append(0)  # Uptrend
        elif all(next_candles['Close'] < next_candles['Open']):
            labels.append(1)  # Downtrend
        else:
            labels.append(2)  # Neutral
    # Pad last 5 rows with Neutral (or drop)
    labels += [2]*5
    return labels

df['Trend_5'] = generate_labels(df)

# Step 5: Feature Selection
# Remove non-numeric or target columns
feature_columns = df.columns.difference(['Date Time', 'Trend_5', 'KAMA COLOR'])
X = df[feature_columns].values
y = df['Trend_5'].values

# Step 6: Scaling Features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 7: Create Sequences for LSTM
def create_sequences(X, y, time_steps=20):
    X_seq, y_seq = [], []
    for i in range(len(X) - time_steps):
        X_seq.append(X[i:i+time_steps])
        y_seq.append(y[i+time_steps])
    return np.array(X_seq), np.array(y_seq)

TIME_STEPS = 20
X_seq, y_seq = create_sequences(X_scaled, y, TIME_STEPS)

# Step 8: Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X_seq, y_seq, test_size=0.2, shuffle=False
)

# Step 9: Build LSTM Model
model = Sequential([
    LSTM(64, input_shape=(X_train.shape[1], X_train.shape[2]), return_sequences=True),
    Dropout(0.2),
    LSTM(32),
    Dropout(0.2),
    Dense(32, activation='relu'),
    Dense(3, activation='softmax')  # 3 classes: Uptrend, Downtrend, Neutral
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.summary()

# Step 10: Train Model
early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

history = model.fit(
    X_train, y_train,
    epochs=50,
    batch_size=32,
    validation_split=0.2,
    callbacks=[early_stop],
    verbose=1
)

# Step 11: Evaluate Model
y_pred = np.argmax(model.predict(X_test), axis=1)

print("Classification Report:\n", classification_report(y_test, y_pred))
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Up', 'Down', 'Neutral'], yticklabels=['Up', 'Down', 'Neutral'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

# Step 12: Plot Training History
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title('Loss Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.subplot(1,2,2)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Val Accuracy')
plt.title('Accuracy Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

# Step 13: Predict on New Data
# Example: predict trend for last TIME_STEPS candles
last_sequence = X_scaled[-TIME_STEPS:].reshape(1, TIME_STEPS, X_scaled.shape[1])
pred = np.argmax(model.predict(last_sequence), axis=1)
trend_dict = {0: "Uptrend", 1: "Downtrend", 2: "Neutral"}
print("Predicted trend for next 5 candles:", trend_dict[pred[0]])
```
